```python
def arrow_for_change(val):
    val = round(val, 2)
    if (val<0):
        val = -1*val
        return '↓ '+str(val)+'%'
    if (val>0):
        return '↑ '+str(val)+'%'
    return '0.0%'
```

```python
merged['Thickness Change'] = merged['Thickness Change'].apply(arrow_for_change)
merged['Area Change'] = merged['Area Change'].apply(arrow_for_change)
```
