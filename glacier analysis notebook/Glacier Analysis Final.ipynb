### Area Change

```python
# The area of Nepal
AREA_OF_NEPAL_SQ_KM = 147181
```

```python
area_of_glaciers_in_1980 = gl_1980.Area_SqKm.sum()
area_of_glaciers_in_2010 = gl_2010.Area_SqKm.sum()
```

```python
area1980_in_percentage = 100*area_of_glaciers_in_1980/AREA_OF_NEPAL_SQ_KM
area2010_in_percentage = 100*area_of_glaciers_in_2010/AREA_OF_NEPAL_SQ_KM
```

```python
print ('Between 1980 to 2010, the overall land area of Nepal covered by glaciers decreased from '+str(area1980_in_percentage)+' to '+ str(area2010_in_percentage))
```

Since there are more glaciers in 2010 dataset (probably because they were discoverd later), the statistics above is a little misleading. 

```python
gl_merged = pd.merge(gl_2010, gl_1980[['GLIMS_ID', 'Area_SqKm', 'Thickness']], how='inner', on='GLIMS_ID')
```

```python
print (f'Only {gl_merged.shape[0]} glaciers are common between 1980 and 2010.\n{gl_2010.shape[0]-gl_merged.shape[0]} are new glaciers in 2010 dataset. \n{gl_1980.shape[0]-gl_merged.shape[0]} are missing from 1980 dataset.')
```

```python
gl_merged.rename(inplace=True, columns={"Area_SqKm_x": "Area_2010", "Area_SqKm_y": "Area_1980", "Thickness_x": "Thickness_2010", "Thickness_y": "Thickness_1980"})
```

```python
gl_merged['AreaDecreasePercentage'] = 100*(gl_merged['Area_1980'] - gl_merged['Area_2010'])/gl_merged['Area_1980']
```

```python
gl_merged['ThicknessDecreasePercentage'] = 100*(gl_merged['Thickness_1980'] - gl_merged['Thickness_2010'])/gl_merged['Thickness_1980']
```

```python
gl_merged[['AreaDecreasePercentage', 'ThicknessDecreasePercentage']].head()
```

```python
gl_merged.plot()
```

## Most Vulnarable Basin and Sub-Basin based on Glaciers data available on both datasets

```python
gl_merged.Basin.value_counts()
```

```python
gl_basin = gl_merged[['GLIMS_ID', 'Basin', 'Area_1980', 'Area_2010', 'Thickness_1980', 'Thickness_2010']].groupby('Basin').sum()
```

```python
gl_basin
```

```python
gl_basin['Thickness_Delta'] = (gl_basin['Thickness_1980']-gl_basin['Thickness_2010'])/gl_basin['Thickness_1980']
gl_basin['Area_Delta'] = (gl_basin['Area_1980']-gl_basin['Area_2010'])/gl_basin['Area_1980']
```

```python
gl_basin.sort_values(by=['Thickness_Delta'], ascending=False)
```

```python
gl_basin.sort_values(by=['Area_Delta'], ascending=False)
```

```python
gl_subbasin = gl_merged[['GLIMS_ID', 'Sub_Basin', 'Area_1980', 'Area_2010', 'Thickness_1980', 'Thickness_2010']].groupby('Sub_Basin').sum()
```

```python
gl_subbasin
```

```python
gl_subbasin['Thickness_Delta'] = (gl_subbasin['Thickness_1980']-gl_subbasin['Thickness_2010'])/gl_subbasin['Thickness_1980']
gl_subbasin['Area_Delta'] = (gl_subbasin['Area_1980']-gl_subbasin['Area_2010'])/gl_subbasin['Area_1980']
```

```python
gl_subbasin.sort_values(by=['Thickness_Delta'], ascending=False)
```

```python
gl_subbasin.sort_values(by=['Area_Delta'], ascending=False)
```
